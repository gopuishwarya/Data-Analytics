```python
y=list(df.Age)
plt.boxplot(y)
plt.show()
```

```python
df.groupby(['Gender','Diastolic']).mean()
```

```python
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df.corr(),cbar=True,cmap='Blues')
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x='Age',y='Location',data=df)
plt.xlabel('Age')
plt.ylabel('Location')
```

```python
sns.histplot(df.Gender,bins=2)
```

```python
category_counts =df['SelfAssessedHealthStatus'].value_counts()
```

```python
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
db=df.groupby('Gender')['Diastolic'].sum().to_frame().reset_index()
plt.barh(db['Gender'],db['Diastolic'],color=['red','pink'])
plt.title('Bar chart')
plt.xlabel('Diastolic')
plt.ylabel('Gender')
plt.show()
```
